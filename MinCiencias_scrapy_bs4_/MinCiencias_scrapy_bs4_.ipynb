```python
import urllib.request
from bs4 import BeautifulSoup
bachelor_list = []
product_list = []
iter1, iter2 = 0, 0

while True:
# Abrimos la pagina web, leemos el contenido y asignamos el HTML a la variable.
  try:
    URL = "https://sba.minciencias.gov.co/Buscador_HojasDeVida/busqueda?q=*&pagenum=@&start=@&type=load&lang=es"
    iter1 += 1
    iter2 = 10 * (iter1 - 1)
    URL = URL.replace('@', str(iter1) , 1)
    URL = URL.replace('@', str(iter2) , 1)
    html = urllib.request.urlopen(URL).read().decode()

#Transformamos el HTML en un objeto de BeautifulSoup
    from bs4 import BeautifulSoup
    soup = BeautifulSoup(html)

#Extraemos la cantidad de investigadores en la pagina

    tags_name = soup.find_all(id='u217-4-Crg')
#Contador de Investigadores
    count_inv = 0
    for t in range(len(tags_name)):

#Extraemos link de hojas de vida
      id_cv = "link_res_" + str(t)
      tags_cv = soup.find_all('a',id=id_cv)
      
#Reemplazamos los caracteres para acceder al link de la hoja de vida
      new_url = tags_cv[0].get('href')
      new_url = new_url.split('%3F')
      new_url[0]+= '?'
      new_url[1] = new_url[1].split("%3D")
      new_url[1][0] += '='
      new_url = new_url[0] + new_url[1][0] + new_url[1][1]
                
#Accedemos a cada link de hoja de vida
      html2 = urllib.request.urlopen(new_url)

#Trasnformamos el nuevo HTML en un objeto BeautifulSoup
      soup2 = BeautifulSoup(html2, 'html.parser')
      tags2 = soup2.find_all('b',string='Pregrado/Universitario')
      bachelor_list_tags = []
      for tag2 in tags2:
        bachelor_list_tags += [tag2.find_parent(name='td').contents[3].strip()]
      bachelor_list.append(bachelor_list_tags)
##

  
#Extraemos link de productos
      id_pd = "title_res_" + str(t)
      tags_pd = soup.find_all('div',id=id_pd)
      url_pd = tags_pd[0].find('a', href=True).get('href')

#Reemplazamos los caracteres para acceder al link de productos
      url_pd = url_pd.replace('..', "https://sba.minciencias.gov.co")
      url_pd = url_pd.replace(' ', "%20")

      iter3,iter4 = 0, 0

      while True:
# Abrimos la pagina web, leemos el contenido y asignamos el HTML a la variable.
        try:
          url_pd_s1 = url_pd.split("&pagenum=")
          url_pd_s2 = url_pd_s1[1].split("&start=")
          url_pd_s3 = url_pd_s2[1].split("&type=")
          iter3 += 1
          iter4 = 10 * (iter3 - 1)
          url_pd_s1[0] += '&pagenum=' + str(iter3) + '&start=' + str(iter4) + '&type='
          url_pd = url_pd_s1[0] + url_pd_s3[1]

  #Accedemos a cada link de los productos
          html3 = urllib.request.urlopen(url_pd).read().decode()

  #Trasnformamos el nuevo HTML en un objeto BeautifulSoup
          soup3 = BeautifulSoup(html3, 'html.parser')
          if len(soup3.find_all('div',id='u218-5-Crg')) > 0:
            if product_list[count_inv].count([]) == False :
              product_list.append(len(soup3.find_all('div',id='u218-5-Crg')))
            else:
              product_list[count_inv] += len(soup3.find_all('div',id='u218-5-Crg'))
          else:
            product_list.append(0)
            break
        except StopIteration:
          break
      print(bachelor_list)
      print(product_list)

  except StopIteration:
    break
```
